# Analyse des écarts entre deux versions
------------------------------
**Synthèse des écarts entre le 17/06/23 et le 04/07/23:**
- **37 821** lignes ont été supprimées qui correspondent à 
- **13 128** pdc supprimés et 526 ajoutés,
    - dont **11 962** suppressions et 464 ajouts de pdc ne présentant aucun écart
    - et **875** suppressions et 45 ajouts de pdc présentant des écarts
- on ne retrouve le 04/07/23 aucun autre pdc avec un id de station ou bien un id de pdc ou bien une coordonnée ou bien un nom de station correspondant aux 13 128 pdc supprimés

**Exemples de données supprimées:**
4 exemples sont fournis :
- **stations** : FRPD1PIBSCLT, FRFR1PZ8TORQXRCM, FRSPSPSHEL611, FRMAPE000024180
- **pdc** : FRPD1EIBSCLTALF22011, FRFR1EWKHE2, FRSPSESHEL611, FRMAPE000024180461
- **coordonnées** : [0.698738, 47.3358432], [5.2535804, 47.11048942], [1.633191, 48.992067], [1.1391721371977237, 42.98728553935362]
- **nom** : 'Ibis Styles Tours', 'Béziers, Route de Maureilhan', 'AIRE DE ROSNY SUR SEINE SUD', 'Val de Gasogne Gam vert Saint Girons'

```python
from datetime import datetime
import json
from observation import Ilist, Analysis
import pandas as pd

chemin = 'https://raw.githubusercontent.com/loco-philippe/Environmental-Sensing/main/python/Validation/irve/Analyse/'
dates = ['2023-03-03', '2023-04-17', '2023-05-02', '2023-05-24', '2023-06-17', '2023-07-04']
relations = ['index', 'contact_operateur', 'nom_enseigne', 'coordonneesXY', 'adresse_station', 'id_station_itinerance', 
             'nom_station', 'implantation_station', 'nbre_pdc', 'condition_acces', 'horaires', 'station_deux_roues', 
             'id_pdc_itinerance', 'date_maj', 'last_modified']

date = dates[-1]
file1 = 'consolidation-etalab-schema-irve-statique-v-2.2.0-'+date[:4]+date[5:7]+date[8:]+'.csv'
valide1 = 'IRVE_itinerance_valide'+date+'.csv'
residuel1 = 'IRVE_itinerance_residuel'+date+'.csv'
irve_ini_1 = pd.read_csv(chemin + file1,     sep=',', low_memory=False)
irve_val_1 = pd.read_csv(chemin + valide1,   sep=',', low_memory=False)
irve_res_1 = pd.read_csv(chemin + residuel1, sep=',', low_memory=False)
irve_1 = pd.concat([irve_val_1, irve_res_1])
irve_ini_1_id = irve_ini_1['id_pdc_itinerance']
print('nombre de lignes fichier 1 récent: initiales, valides et résiduelles', len(irve_ini_1), len(irve_val_1), len(irve_res_1)) 

date = dates[-2]
file2 = 'consolidation-etalab-schema-irve-statique-v-2.2.0-'+date[:4]+date[5:7]+date[8:]+'.csv'
valide2 = 'IRVE_itinerance_valide'+date+'.csv'
residuel2 = 'IRVE_itinerance_residuel'+date+'.csv'
irve_ini_2 = pd.read_csv(chemin + file2,     sep=',', low_memory=False)
irve_val_2 = pd.read_csv(chemin + valide2,   sep=',', low_memory=False)
irve_res_2 = pd.read_csv(chemin + residuel2, sep=',', low_memory=False)
irve_2 = pd.concat([irve_val_2, irve_res_2])
irve_ini_2_id = irve_ini_2['id_pdc_itinerance']
print('nombre de lignes fichier 2 ancien: initiales, valides et résiduelles', len(irve_ini_2), len(irve_val_2), len(irve_res_2)) 
```

```python
irve_ini_1['new'] = ~irve_ini_1['id_pdc_itinerance'].isin(irve_ini_2_id)
print('nombre de pdc ajoutés : ', irve_ini_1['new'].sum())

irve_ini_2['new'] = ~irve_ini_2['id_pdc_itinerance'].isin(irve_ini_1_id)
print('nombre de pdc supprimés : ', irve_ini_2['new'].sum())
```

```python
irve_res_1['new'] = ~irve_res_1['id_pdc_itinerance'].isin(irve_ini_2_id)
print('nombre de pdc résiduels ajoutés : ', irve_res_1['new'].sum())

irve_res_2['new'] = ~irve_res_2['id_pdc_itinerance'].isin(irve_ini_1_id)
print('nombre de pdc résiduels supprimés : ', irve_res_2['new'].sum())
irve_res_2[irve_res_2['new']].loc[:, relations]
```
