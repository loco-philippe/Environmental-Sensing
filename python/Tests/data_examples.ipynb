## Jeu de test fonctionnel
Exemple de jeu de test

```python
# jeu de tests - diversité des données
liste_obs_mixte = []
for i in range(20):
    liste_obs_mixte += [ob_mixte_2(i, i)]

print('nb observations : ', len(liste_obs_mixte))
print('nb records : ', sum([len(obs) for obs in liste_obs_mixte]))
```

```python
# jeu de tests - données représentatives
liste_obs_tests = []
for i in range(10):
    liste_obs_tests += [ob_mesure(res=i, jour=i, mois=1, lieu=0, prop=0)]
    liste_obs_tests += [ob_mesure(res=10+i, jour=i, mois=1, lieu=i+1, prop=1)]
    liste_obs_tests += [ob_mesure(res=100+i, jour=i, mois=1, lieu=i+1, prop=i)]
for i in range(10):
    liste_obs_tests += [ob_signal(jour=i, mois=1, lieu=i, nuis=i, intens=i)]
liste_obs_tests += [ob_fixe(dj=2, nh=24), ob_fixe(dj=3, nh=24)]
liste_obs_tests += [ob_mob_1(d=0, nval=10), ob_mob_1(d=1, nval=10)]
liste_obs_tests += [ob_mobile(d=1, nval=10), ob_mob_1(d=2, nval=10)]
liste_obs_tests += [ob_multi(dj=3, nh=24, nloc=10), ob_multi(dj=4, nh=24, nloc=10)]
liste_obs_tests += [ob_dalle(dj=4, nbd=10, lg=0.3), ob_dalle(dj=4, nbd=10, lg=0.5)]
liste_obs_tests += [ob_multi_dalle(dj=5, nh=5, nbd=10, lg=0.3), 
                    ob_multi_dalle(dj=6, nh=5, nbd=10, lg=0.4)]

print('nb observations : ', len(liste_obs_tests))
print('nb records : ', sum([len(obs) for obs in liste_obs_tests]))
```

```python
# jeu de tests - performance
coef = 2
liste_obs_tests = []
for i in range(10 * coef):
    liste_obs_tests += [ob_mesure(res=i, jour=i, mois=1, lieu=0, prop=0)]
    liste_obs_tests += [ob_mesure(res=10+i, jour=i, mois=1, lieu=i+1, prop=1)]
    liste_obs_tests += [ob_mesure(res=100+i, jour=i, mois=1, lieu=i+1, prop=i)]
for i in range(10 * coef):
    liste_obs_tests += [ob_signal(jour=i, mois=1, lieu=i, nuis=i, intens=i)]
liste_obs_tests += [ob_fixe(dj=2, nh=24 * coef), ob_fixe(dj=3, nh=24 * coef)]
liste_obs_tests += [ob_mob_1(d=0, nval=10 * coef), ob_mob_1(d=1, nval=10 * coef)]
liste_obs_tests += [ob_mobile(d=1, nval=10 * coef), ob_mob_1(d=2, nval=10 * coef)]
liste_obs_tests += [ob_multi(dj=3, nh=24 * coef, nloc=10), ob_multi(dj=4, nh=24 * coef, nloc=10)]
liste_obs_tests += [ob_dalle(dj=4, nbd=10 * coef, lg=0.3 / coef), ob_dalle(dj=4, nbd=10 * coef, lg=0.5 / coef)]
liste_obs_tests += [ob_multi_dalle(dj=5, nh=5, nbd=10 * coef, lg=0.3 / coef), 
                    ob_multi_dalle(dj=6, nh=5, nbd=10 * coef, lg=0.4 / coef)]

print('nb observations : ', len(liste_obs_tests))
print('nb records : ', sum([len(obs) for obs in liste_obs_tests]))
```

```python
from time import time
list_json_tests = []
t00= time()
info = True
detail = False
for i in range(40 * coef):
    list_json_tests.append(liste_obs_tests[i].to_obj(modecodec='dict', json_info=info, json_info_detail=detail))   
print(time()-t00)
for i in range(40 * coef, len(liste_obs_tests)):
    t0 = time()
    list_json_tests.append(liste_obs_tests[i].to_obj(modecodec='dict', json_info=info, json_info_detail=detail))
    print(time()-t0)
print('coef : ', coef)
print('total : ', time() - t00)
```
